```python
from Statistics2 import *
```

```python
vec = VecStat(730)
for d in data1024:
    vec.accum(d)
result = vec.compute(10)
```

```python
(countV,meanV)=vec.V.compute()
```

```python
(countC,meanC)=vec.Cov.compute()
```

```python
cov=meanC-outer(meanV,meanV)
```

```python
(eigvalues,eigvectors)=linalg.eig(cov)
```

```python
order=argsort(-abs(eigvalues))
```

```python
eigvectors = eigvectors.transpose()
eigvectors=eigvectors[order]   # order eigenvectors
print eigvectors
```
